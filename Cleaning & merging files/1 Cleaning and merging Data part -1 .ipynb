```python
# Loading files individaully
name_data = "C:\\Users\\vishd\Desktop\\reetu\\Springboard course\\Data wrangling IMDB dataset\\1name_data.tsv"
title_akas = "C:\\Users\\vishd\Desktop\\reetu\\Springboard course\\Data wrangling IMDB dataset\\2title_akas.tsv"
title_basics = "C:\\Users\\vishd\Desktop\\reetu\\Springboard course\\Data wrangling IMDB dataset\\3title_basics.tsv"
title_crew = "C:\\Users\\vishd\Desktop\\reetu\\Springboard course\\Data wrangling IMDB dataset\\4title_crew.tsv"
title_episode = "C:\\Users\\vishd\Desktop\\reetu\\Springboard course\\Data wrangling IMDB dataset\\5title_episode.tsv"
tilte_principle = "C:\\Users\\vishd\Desktop\\reetu\\Springboard course\\Data wrangling IMDB dataset\\6tiltle_principle.tsv"
title_ratings = "C:\\Users\\vishd\Desktop\\reetu\\Springboard course\\Data wrangling IMDB dataset\\7title_ratings.tsv"
```

```python
title_akas = pd.read_csv(title_akas, sep='\t')                                         
title_akas.head()
```

```python
title_akas_dropped = title_akas.drop_duplicates('titleId',keep=False)                    
title_akas_dropped.head()
```

```python
title_basics = pd.read_csv(title_basics, sep='\t')
title_basics.head(2)
#title_basics.shape
```

# Merging - 1 

```python
join1 = title_akas_dropped.merge(title_basics, left_on='titleId', right_on='tconst', how='outer')
join1.shape                                                                       
```

```python
#join1['isOriginalTitle'].nunique(dropna = True)                 # Checking unique values in columns and dropping them) 
#[languages, types, attributes, tconst, isAdult, primaryTitle, originalTitle, 'endyear(with 96)',runtimeMinutes(677), 'ordering(1)' ]  

join1 =join1.drop(['language','types', 'attributes', 'tconst', 
            'isAdult', 'primaryTitle', 'originalTitle',
            'isOriginalTitle', 'endYear', 'runtimeMinutes','ordering'], axis=1)
                                                                
join1.head()
join1.shape
```

```python
title_crew = pd.read_csv(title_crew, sep='\t')
title_crew.head(2)
#title_crew.shape
```

# Merging 2

```python
join2 = join1.merge(title_crew, left_on='titleId', right_on='tconst', how='outer')        
join2.head()
```

```python
join2 =join2.drop(['tconst'], axis=1)
#join2.head()
join2.shape
```

```python
title_episode = pd.read_csv(title_episode, sep='\t')
title_episode.head(2)
#title_episode.shape
```

# Merging 3

```python
Join3 = join2.merge(title_episode, left_on='titleId', right_on='tconst', how='outer') # merging title_episode with previous df
Join3.head()
```

```python
Join3 =Join3.drop(['tconst'], axis=1)
Join3.head()
#Join3.shape
```

```python
title_ratings = pd.read_csv(title_ratings, sep='\t')
title_ratings.head(2)
```

# Merging - 4

```python
Join4 = Join3.merge(title_ratings, left_on='titleId', right_on='tconst', how='outer') 
Join4.head()
```

```python
Join4 =Join4.drop(['tconst'], axis=1)
Join4.head()
#print(Join4.shape)
```
